```
from pyspark.sql.functions import concat_ws
from pyspark.sql.functions import lit, col
cigna_file_path = '/mnt/hype-ml.compassphs.com/bcbs_Med_Claim_LineItem.csv'
health_df = health_condition(ICD10_file_path, comparision_dict)
def health_Analysis(cigna_file_path, health_df):
  cigna_df = spark.read.csv(cigna_file_path, sep='|',header=True).select('Diagnosis_1_Key', 'Diagnosis_2_Key', 'Diagnosis_3_Key', 'Diagnosis_4_Key', 
                                                                          'Diagnosis_5_Key', 'Diagnosis_6_Key', 'Amount_Allowed', 'Start_Date', 'Employer_Key')
  cigna_df_col_combined = cigna_df.withColumn('Diagnosis_1_Key', concat_ws(',', *[col for col in cigna_df.columns if col.startswith('Diagnosis')]))

  cigna_df_drop_col = cigna_df_col_combined.drop(*[col for col in cigna_df_col_combined.columns if col.startswith('Diagnosis') and col != 'Diagnosis_1_Key'])

  cigna = cigna_df_drop_col.where((col("Start_Date") >= "2018-01-01") & (col("Start_Date") <= "2022-12-31"))
  dedup_cigna = cigna.dropDuplicates()
  cigna_filled = dedup_cigna.fillna(0) # fill null values
  filtered_dfs = [] # initialize empty list to store DataFrames
  for k, v in health_df.items():
    filtered_df = cigna_filled.filter(cigna_filled['Diagnosis_1_Key'].isin(v))\
                               .select('Diagnosis_1_Key', 'Amount_Allowed', 'Start_Date','Employer_Key')
    filtered_df = filtered_df.withColumn('condition', lit(k))
    filtered_dfs.append(filtered_df) # append filtered DataFrame to list
  filtered_df = spark.createDataFrame([], schema=filtered_dfs[0].schema) # create empty DataFrame with schema
  for df in filtered_dfs:
    filtered_df = filtered_df.union(df) # concatenate filtered DataFrames
  return filtered_df
```
